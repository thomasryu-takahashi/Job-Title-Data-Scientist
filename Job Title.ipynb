```python
import pandas as pd
import nltk 
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from googletrans import Translator
from nltk import sent_tokenize, word_tokenize
```

```python
#importa a base
df = pd.read_csv('./Data_science_linkedin_PORT.csv')
df.head()
```

```python
#põe tudo em minusculo
df['descricao_port'] = df['descricao_port'].str.lower()
df.head()
```

```python
#remove digitos
df['descricao_port'] = df['descricao_port'].str.replace(r'[0-9]', '')
df.head()
```

```python
#remove "erros de importação"
df['descricao_port'] = df['descricao_port'].str.replace('\r', ' ')
df['descricao_port'] = df['descricao_port'].str.replace('\n', ' ')
df.head()
```

```python
#remove stopwords
stopwords = set(nltk.corpus.stopwords.words('portuguese'))
def remove_stopwords(text):
    return [word for word in text.split() if word not in stopwords]
```

```python
df['norm'] = df['descricao_port'].apply(remove_stopwords)
```

```python
def rejoin_words(row):
    my_list = row['norm']
    joined_words = ( " ".join(my_list))
    return joined_words

df['norm_1'] = df.apply(rejoin_words, axis=1)
```

```python
df.head()
```

```python
#Retira acentuação
df['norm_1'] = df['norm_1'].str.normalize('NFKD').str.encode('ascii', 'ignore').str.decode('utf-8')
df.head()
```

```python
#remoção de pontuação
import string

translation_table = str.maketrans({key: ' ' for key in string.punctuation}) 

def remove_punctuation(text):
    return text.translate(translation_table)
```

```python
df['norm_1'] = df['norm_1'].apply(remove_punctuation)
df.head()
```

```python
df_normalizada = df[['Empresa','norm_1']]
```

```python
#exporta a base para importar para o R
df_normalizada.to_csv('./normalizada_PORT.csv', index=False)
```
